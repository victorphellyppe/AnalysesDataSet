Analise de filmes do brasil de 2009 a 2017
arquivo -> 2120_0.xlsx

```python
#importando bibliotecas
import pandas as pd
#configurando matplotlib
%matplotlib inline
```

```python
#criando o dataframe 
#pulando as duas primeiras linhas
df = pd.read_excel('2120_0.xlsx',skiprows=2,skipfooter=16)
#Exibindo as 5 primeiras linhas do dataframe
df.head(3)
```

```python
#Exibindo o final do dataframe
df.tail(20)
```

```python
#Verificar tipos das colunas    object = ou é texto ou não consiguiu reconhcer
df.dtypes
```

```python
#Converter o campo datadelancamento em data
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
#Vamos tentar identificar o problema da coluna datadelancamento

df['Data de lançamento'].value_counts()
```

```python
#Exibir alguns filmes que estao com data relancamento

df[df['Data de lançamento'] == 'Relançamento'].head()
```

##Decisão sobre dados incorretos/faltantes/divergentes

```python
#Tamanho do dataframe e o formato é (linhas,colunas)
df.shape
```

```python
# Criar um novo dataframe sem os relançamentos(ao inves de apagar)
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
#Vamos verificar os tipos das colunas
df_novo.dtypes
```

```python
#Convertendo a coluna data de lançamento para datetime
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.dtypes
```

```python
#quais os anos tiveram mais filmes lançados

df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
#Plotar um gráfico de barras com dados de uma coluna
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```
